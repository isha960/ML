```python
import pandas as pd
df1 = pd.read_csv('2020.csv')
df1
```

```python
df2 = pd.read_csv('2019.csv')
df2
```

```python
df3 = pd.read_csv('2018.csv')
df3
```

```python
df4 = pd.read_csv('2017.csv')
df4
```

```python
df5 = pd.read_csv('2016.csv')
df6 = pd.read_csv('2015.csv')
```

```python
df1['year']=2020
df2['year']=2019
df3['year']=2018
df4['year']=2017
df5['year']=2016
df6['year']=2015
```

```python
df1.columns = map(str.lower, df1.columns)
df2.columns = map(str.lower, df2.columns)
df3.columns = map(str.lower, df3.columns)
df4.columns = map(str.lower, df4.columns)
df5.columns = map(str.lower, df5.columns)
df6.columns = map(str.lower, df6.columns)
```

```python
final=pd.concat([df1,df2,df3,df4,df5,df6])
final
```

```python
final['social support']=final['family']
final.drop(columns='family',inplace=True)
final
```

```python
final.sort_values(["happiness score"], axis=0, 
                 ascending=False, inplace=True) 
last=final.tail(10)
first=final.head(10)
first
```

```python
df7= first.head(1)
df7= df7.stack()
df7
df7.drop(df7.tail(1).index,inplace=True)
df7
df7.drop(df7.head(3).index,inplace=True)
df7
```

```python
df7 = pd.DataFrame(df7,columns=['finland'])
df7.reset_index
df7
```
